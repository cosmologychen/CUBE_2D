```python
def loadfield2d(fn):
    """
    fn: filename
    Returns:
    a: 2D field
    """
    fid = open(fn, 'rb')
    # fid = open(fn, 'rb', 'big')  # big endian
    p1 = np.fromfile(fid, dtype=np.float32)  # real*4
    fid.close()
    n = round(len(p1) ** (1/2))
    if (n**2 != len(p1)):
        print('shape no mach')
        print(n,n**2,len(p1))
        return 0,0
    a = np.reshape(p1, (n, n),order='F')
    # print(fn,n)
    return a,n

def plt_delta(name,box='400',Redshift='5.000',cl=[-5,5],cmap='bwr',save=False,times=1):
    delta,n = loadfield2d(f'{Path}{Redshift}_{name}.bin')
    plt.figure(figsize=(4,3),dpi=300)
    plt.imshow(delta*times, interpolation='none',cmap=cmap)
    # plt.imsave(f'200_{name}.png', delta.T, cmap=cmap, origin='lower')
    plt.title(f'{box}_{name}_{Redshift}')
    if (cl != 'auto'):
        plt.clim(cl[0],cl[1])
    # plt.clim(-1,3)
    if (times == 1):
        plt.colorbar(label=f'{name}')
    else:
        plt.colorbar(label=f'{times:.1e} {name}')
    print(np.max(delta),np.min(delta))
    if save:
        os.makedirs(f'{Path}/fig', exist_ok=True)
        plt.savefig(f'{Path}/fig/{box}_{name}_{Redshift}.pdf', bbox_inches='tight')
        plt.close()
    else:
        plt.show()
        
def loadpower(filename):
    n_row_xi = 10
    fid = open(filename, 'rb')  # 以二进制模式打开文件

    xi = np.fromfile(fid, dtype='float32')  # 读取数据到一维数组
    fid.close()

    xi = np.reshape(xi, (int(len(xi) / n_row_xi), n_row_xi))  # 重新形状为二维数组
    ksim = xi[:, 1]  # 提取第二列作为 ksim

    return ksim, xi
```

```python
def load2dxpos(fn):
    fid = open(fn, 'rb')
    p1 = np.fromfile(fid, dtype=np.float32)
    fid.close()
    n = round(len(p1)/2)
    if (n*2 != len(p1)):
        print('shape no mach')
        print(n,n**2,len(p1))
        return 0,0
    a = np.reshape(p1, (2, n),order='F')
    # print(fn,n)
    return a,int(np.sqrt(n))
```

```python
# DTFE 三角剖分结果可视化
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def read_triangles(filename):
    # 首先读取原始二进制数据
    data = np.fromfile(filename, dtype=np.float32)
    
    # 每个三角形有 3 个点，每个点 2 个坐标
    # 所以每个三角形 = 3 × 2 = 6 个数
    if data.size % 6 != 0:
        raise ValueError("文件大小不是 (n × 3 × 2)的倍数")

    ntri = data.size // 6
    triangles = data.reshape((ntri, 3, 2))
    return triangles

# 示例用法
triangles = read_triangles("./utilities/triangles.bin")
downsample = 5
triangles_sownsample = triangles[::downsample,:,:]
fig, ax = plt.subplots(figsize=(16, 16))

for tri in tqdm(triangles_sownsample, desc="绘制三角形"):
    polygon = patches.Polygon(tri, closed=True, edgecolor='black', facecolor='none', linewidth=0.4)
    ax.add_patch(polygon)

ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('DTFE Triangulation')
plt.xlim(0,1024)
plt.ylim(0,1024)
plt.tight_layout()
plt.show()
```

```python


# === Step 5: 密度过滤（距离<100内只保留 rho_D 更小的点）===
coords = np.column_stack(np.where(mask))
tree = cKDTree(coords)
keep = np.ones(len(coords), dtype=bool)


r = 200
roi = r/2
for i in tqdm(range(len(coords))):
    if not keep[i]:
        continue
    xi, yi = coords[i]
    neighbors = tree.query_ball_point(coords[i], r=r)
    avg_rho_i = np.mean(mean_DE_smooth[int(max(0, xi - roi)):int(xi + roi + 1), int(max(0, yi - roi)):int(yi + roi + 1)])

    for j in neighbors:
        if i == j or not keep[j]:
            continue
        xj, yj = coords[j]
        avg_rho_j = np.mean(mean_DE_smooth[int(max(0, xj - roi)):int(xj + roi + 1), int(max(0, yj - roi)):int(yj + roi + 1)])

        if avg_rho_i > avg_rho_j:
            keep[i] = False
            break
        else:
            keep[j] = False

# === Step 6: 提取并可视化保留的极小值点 ===
filtered_coords = coords[keep]
min_y, min_x = filtered_coords[:, 0], filtered_coords[:, 1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binned_statistic

def compute_power_spectrum(delta, box_size=1.0, bins=50):
    assert delta.shape[0] == delta.shape[1], "delta 必须是方形网格"
    print(np.mean(delta))
    n = delta.shape[0]
    # 2D FFT 并移到中心
    delta_k = np.fft.fftshift(np.fft.fft2(delta))/n**2
    
    # 计算功率谱密度
    power_2d = np.abs(delta_k)**2

    # 构造 k-space 网格
    kfreq = np.fft.fftshift(np.fft.fftfreq(n, d=box_size/n))  # 频率单位: 1/box_size
    kx, ky = np.meshgrid(kfreq, kfreq, indexing='ij')
    k_mag = np.sqrt(kx**2 + ky**2)

    # 将2D功率谱角平均为1D
    k_flat = k_mag.flatten()
    p_flat = power_2d.flatten()

    k_bins = np.linspace(0, np.max(k_mag), bins + 1)
    Pk_1d, _, _ = binned_statistic(k_flat, p_flat, statistic='mean', bins=k_bins)

    # k_bin 的中心值
    k_bin_centers = 0.5 * (k_bins[1:] + k_bins[:-1])

    return k_bin_centers, Pk_1d
```
