```python
def loadfield2d(fn):
    """
    fn: filename
    Returns:
    a: 2D field
    """
    fid = open(fn, 'rb')
    # fid = open(fn, 'rb', 'big')  # big endian
    p1 = np.fromfile(fid, dtype=np.float32)  # real*4
    fid.close()
    n = round(len(p1) ** (1/2))
    if (n**2 != len(p1)):
        print('shape no mach')
        print(n,n**2,len(p1))
        return 0,0
    a = np.reshape(p1, (n, n),order='F')
    # print(fn,n)
    return a,n

def load2dxpos(fn):
    fid = open(fn, 'rb')
    p1 = np.fromfile(fid, dtype=np.float32)
    fid.close()
    n = round(len(p1)/2)
    if (n*2 != len(p1)):
        print('shape no mach')
        print(n,n**2,len(p1))
        return 0,0
    a = np.reshape(p1, (2, n),order='F')
    # print(fn,n)
    return a,int(np.sqrt(n))

def load2dvoid(fn):
    with open(fn, 'rb') as f:
        print(fn)
        data = np.fromfile(f, dtype=np.float32)

        n_void = int(len(data)/9)
        print(f"文件中的空洞数量: {n_void}")
        a = data.reshape(9, n_void, order='F')
        print(a.shape)
        center =  a[0:2,:]
        radii =  a[2,:]
        # triangles = a[3:,:]((2,3,n_void))
        triangles = np.zeros((n_void,3, 2))
        triangles[:, 0, :] = a[3:5, :].T  # 点a
        triangles[:, 1, :] = a[5:7, :].T  # 点b
        triangles[:, 2, :] = a[7:9, :].T  # 点c

    return center, radii, triangles

def load2dvoid_r(fn):
    with open(fn, 'rb') as f:
        print(fn)
        data = np.fromfile(f, dtype=np.float32)

        n_void = int(len(data)/9)
        print(f"文件中的空洞数量: {n_void}")
        a = data.reshape(9, n_void, order='F')
        print(a.shape)
        center =  a[0:2,:]
        radii =  a[2,:]
        # triangles = a[3:,:]((2,3,n_void))
        triangles = np.zeros((n_void,3, 2))
        triangles[:, 0, :] = a[3:5, :].T  # 点a
        triangles[:, 1, :] = a[5:7, :].T  # 点b
        triangles[:, 2, :] = a[7:9, :].T  # 点c

    return center, radii, triangles
```

```python
def read_fortran_stream_file(filename):
    """
    读取Fortran流访问格式的文件
    
    参数:
        filename: 文件名
        
    返回:
        包含所有读取数据的列表
    """
    data_list = []
    
    with open(filename, 'rb') as f:
        # 读取文件开头的n_void值
        n_void_bytes = f.read(4)
        n_void = struct.unpack('i', n_void_bytes)[0]
        print(f"文件开头的n_void值: {n_void}")
        
        # 循环读取每个三角形的数据
        while True:
            try:
                # 读取center(2个浮点数), radius(1个浮点数), a,b,c(各2个浮点数)
                # 总共 2+1+2+2+2 = 9个浮点数
                float_data = f.read(9 * 4)
                if not float_data:
                    break
                    
                # 解包浮点数数据
                floats = struct.unpack('9f', float_data)
                x_center = floats[0:2]
                radius = floats[2]
                a = floats[3:5]
                b = floats[5:7]
                c = floats[7:9]
                
                # 读取np_void值
                np_void_bytes = f.read(8)
                np_void = struct.unpack('q', np_void_bytes)[0]
                # print(center, radius, np_void)
                
                # 读取np_void个整数(ip值)
                ip_list = []
                qpos_list = []
                for _ in range(np_void):
                    ip_bytes = f.read(8)
                    ip = struct.unpack('q', ip_bytes)[0]
                    ip_list.append(ip)
                    ip_bytes = f.read(8)
                    q = struct.unpack('ff', ip_bytes)
                    qpos_list.append(q)
                q_center_bytes = f.read(8)
                q_center = struct.unpack('ff', q_center_bytes)
                
                # 读取最后的0值
                zero_bytes = f.read(8)
                zero = struct.unpack('q', zero_bytes)[0]
                if zero != 0:
                    print(len(data_list),np_void,"Error: The last value is not zero.",zero)
                    return
                
                # 将数据添加到列表
                data_list.append({
                    'x_center': x_center,
                    'q_center': q_center,
                    'radius': radius,
                    'triangle': np.array([a,b,c]),
                    'np_void': np_void,
                    'ip_list': ip_list,
                    'qp_list': qpos_list,
                })
                
            except Exception as e:
                print(f"读取文件时出错: {e}")
                break
            
    #data_list按照np_void从大到小排序
    data_list.sort(key=lambda x: x['radius'], reverse=True)

    
    return data_list


def read_void_file(filename):
    halos = []
    with open(filename, 'rb') as f:
        # 讀取文件頭部信息
        header = f.read(12)  # 假設b_link, np_halo_min, np_head各佔4字節
        b_link, np_halo_min_file, np_head = struct.unpack('ffi', header)
        print(f'b_link: {b_link}, np_halo_min: {np_halo_min_file}, np_head: {np_head}')
        
        for _ in range(np_head):
            # 讀取halo粒子數
            void_np_bytes = f.read(4)
            if not void_np_bytes:
                print("缺乏halo_np，文件結束")
                break
            void_np = struct.unpack('i', void_np_bytes)[0]
            
            if void_np == 0:  # 文件結束標記
                print("halo_np为0，文件結束")
                break
            
            # 讀取halo中心位置
            xp_mean_bytes = f.read(8)  # 假設兩個浮點數，每個4字節
            xp_mean = struct.unpack('ff', xp_mean_bytes)
            
            # 讀取halo中所有粒子的位置
            particles = []
            for _ in range(void_np):
                pos_bytes = f.read(8)
                pos = struct.unpack('ff', pos_bytes)
                particles.append(pos)
            
            # 存儲halo信息
            halos.append({
                'np': void_np,
                'center': xp_mean,
                'particles': np.array(particles)
            })
        

    halos.sort(key=lambda x: x['np'], reverse=True)
    
    return halos, b_link, np_halo_min_file, np_head
```

```python
import numpy as np
import struct

def read_halo_file(filename, np_halo_min):
    halos = []
    with open(filename, 'rb') as f:
        # 讀取文件頭部信息
        header = f.read(12)  # 假設b_link, np_halo_min, np_head各佔4字節
        b_link, np_halo_min_file, np_head = struct.unpack('ffi', header)
        
        for _ in range(np_head):
            # 讀取halo粒子數
            halo_np_bytes = f.read(4)
            if not halo_np_bytes:
                print("缺乏halo_np，文件結束")
                break
            halo_np = struct.unpack('i', halo_np_bytes)[0]
            
            if halo_np == 0:  # 文件結束標記
                print("halo_np为0，文件結束")
                break
            
            # 讀取halo中心位置
            xp_mean_bytes = f.read(8)  # 假設兩個浮點數，每個4字節
            xp_mean = struct.unpack('ff', xp_mean_bytes)
            
            # 讀取halo中所有粒子的位置
            particles = []
            for _ in range(halo_np):
                pos_bytes = f.read(8)
                pos = struct.unpack('ff', pos_bytes)
                particles.append(pos)
            
            # 存儲halo信息
            if halo_np > np_halo_min:
                halos.append({
                    'np': halo_np,
                    'center': xp_mean,
                    'particles': np.array(particles)
                })
    
    return halos, b_link, np_halo_min_file, np_head

# 使用示例
filename = f'{Path}halo.bin'
np_halo_min = 10  # 最小halo粒子數閾值
halos, b_link, np_halo_min_file, np_head = read_halo_file(filename, np_halo_min)

# 輸出結果
print(f"b_link: {b_link}")
print(f"np_halo_min: {np_halo_min_file}")
print(f"np_head: {np_head}")
print(f"halo_np_max: {max(halo['np'] for halo in halos)}")
# for i, halo in enumerate(halos):
#     print(f"Halo {i+1}:")
#     print(f"  Particle count: {halo['np']}")
#     print(f"  Center position: {halo['center']}")
#     print(f"  First 5 particles: {halo['particles'][:5]}")
```

```python
from scipy.spatial import cKDTree

def find_close_halos_kdtree(halos, distance_threshold=2.0):
    """
    使用KD树查找距离小于阈值的halo对，并检查粒子间最小距离
    
    参数:
        halos: 包含所有halo信息的列表
        distance_threshold: 中心距离阈值，默认为2.0
        
    返回:
        包含所有符合条件halo对的列表，每个元素为(halo1_index, halo2_index, center_distance, min_particle_distance)
    """
    # 提取halo中心位置
    centers = np.array([halo['center'] for halo in halos])
    
    # 构建KD树
    tree = cKDTree(centers)
    
    # 查找中心距离小于阈值的halo对
    pairs = tree.query_pairs(distance_threshold)
    
    close_pairs = []
    for i, j in pairs:
        # 计算中心距离
        center_distance = np.linalg.norm(centers[i] - centers[j])
        
        # 提取两个halo的粒子位置
        particles_i = halos[i]['particles']
        particles_j = halos[j]['particles']
        
        min_particle_distance = 1
        for particle_i, particle_j in zip(particles_i, particles_j):
            # 计算粒子间距离
            particle_distance = np.linalg.norm(particle_i - particle_j)
            if (particle_distance <0.02) :
                print(particle_i,particle_j)
            min_particle_distance = min(min_particle_distance,particle_distance)
        
        # 存储结果
        if (min_particle_distance <0.02):
            print(i, j, center_distance, min_particle_distance)
    
    return close_pairs

# 使用示例
distance_threshold = 1.0
close_pairs = find_close_halos_kdtree(halos, distance_threshold)

# 输出结果
print(f"Found {len(close_pairs)} halo pairs with center distance < {distance_threshold}:")
for pair in close_pairs:
    i, j, center_dist, min_part_dist = pair
    print(f"Halo {i} and Halo {j}:")
    print(f"  Center distance: {center_dist:.4f}")
    print(f"  Min particle distance: {min_part_dist:.4f}")
    print(f"  Halo {i} particle count: {halos[i]['np']}")
    print(f"  Halo {j} particle count: {halos[j]['np']}")
    print("---")
```
